```python
from locdata.dataset.geojson import GeoJsonDataset
```

```python
source = "../data/osm_nodes_amnenities_dc_bbox.geojson"
center = {"latitude": 38.882843, "longitude":-76.997509}
filter_values = {"amenity":["restaurant","fast_food", "cafe",'bar',"pub"]}
```

```python
gds = GeoJsonDataset(source, filter_values=filter_values)

```

```python
# View Amenity Columns
gds.df['properties.amenity'].value_counts().sort_values(ascending=False).head(50)
```

```python
gds.df.columns
```

```python
#quick visualization on map of raw data

m = Map(location= list(center.values()), zoom_start=11, tiles="cartodbpositron", 
        attr= ' <a href="http://www.openstreetmap.org/copyright">OpenStreetMap</a> contributors  <a href="http://cartodb.com/attributions#basemaps">CartoDB</a>' 
)
mc = MarkerCluster()

for i,row in gds.df.iterrows():
    mk = Marker(location=[row["latitude"],row["longitude"]], popup=row["properties.name"])
    mk.add_to(mc)
    
mc.add_to(m)

m.save('graphics/2_markers_eats.html')
m
```
